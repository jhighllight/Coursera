### Problem Statement
| Size (1000 sqft)     | Price (1000s of dollars) |
| -------------------| ------------------------ |
| 1.0               | 500                      |
| 2.0               | 800                      |

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./deeplearning.mplstyle')
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([500.0, 800.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

### Number of training example  `m`

```python
print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0]
print(f"Number of training example is: {m}")
```

### Training example `x_i, y_i`

```python
i = 1

x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

### Plotting the data

```python
# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='y')
# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollers)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.show()
```

### Model function

Different values of  ùë§ and  ùëè give you different straight lines on the plot.

```python
w = 200
b = 300
print(f"w: {w}")
print(f"b: {b}")
```

Given two data points, compute the value of the function ùëìùë§,ùëè(ùë•(ùëñ)) for each.

```python
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b,)

# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c = 'g', label = 'Our Prediction')

# Plot the data points
plt.scatter(x_train, y_train, marker = 'x', c = 'r', label = 'Actual Values')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the X-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```
